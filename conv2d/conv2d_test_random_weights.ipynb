```python
import tensorflow as tf

# Define hyperparameters
filters = 1  # Number of output filters (feature maps)
kernel_size = (3, 3)  # Size of the convolution kernel
strides = (1, 1)  # Stride of the convolution
padding = "VALID"  # Padding method (to maintain output size)
dilation_rate = (1, 1)  # Dilation rate (controls spacing between kernel elements)

# Create random input with multiple channels
batch_size = 1  # We'll process a single image
input_height = 5  # Example height
input_width = 5  # Example width
input_channels = 3  # Number of input channels (e.g., RGB)
input_tensor = tf.random.normal(shape=(batch_size, input_height, input_width, input_channels))
print(input_tensor)

# Define the convolution layer
conv_layer = tf.keras.layers.Conv2D(
    filters=filters,
    kernel_size=kernel_size,
    strides=strides,
    padding=padding,
    dilation_rate=dilation_rate,
    groups=1,
    activation= None,  # Optional activation function (e.g., ReLU)
    use_bias = None
)

# Build the layer to initialize its weights
conv_layer.build(input_shape=(batch_size, input_height, input_width, input_channels))


# Get the kernel weights
kernel_weights = conv_layer.get_weights()[0]

# Print the kernel weights
print("Kernel Weights:")
print(kernel_weights)


# Apply the convolution
output_tensor = conv_layer(input_tensor)

# Print the input shape
print("Input shape:", input_tensor.shape)

# Print the kernel size
print("Kernel size:", conv_layer.kernel.shape)

# Print the output shape (for verification)
print("Output shape:", output_tensor.shape)

print(output_tensor)
```
